```python
# define undersample strategy
from imblearn.under_sampling import RandomUnderSampler

undersample = RandomUnderSampler(
                    sampling_strategy=1 - df['Class'].value_counts()[1]/df['Class'].value_counts()[0])

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

# fit and apply the transform
X_over, y_over = undersample.fit_resample(X, y)
```

```python
y_over.value_counts()
```

```python
y_over.value_counts().plot(kind='bar', color=['green', 'orange'])
plt.xticks(rotation=0)
plt.xlabel('Class')
plt.ylabel('Number of cases ')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state=0)
```

```python
import seaborn as sns


plt.subplots(figsize=(22,14))
sns.heatmap(X_train.join(pd.DataFrame(y_train)).corr(), annot=True, cmap="YlGnBu", linewidths=.5, fmt='.2f')
```

```python
variables_name = ['V4', 'V11', 'V2', 'V19', 'V14', 'V12', 'V10', 'V16']
sns.pairplot(X_train[variables_name], diag_kind='kde', plot_kws={'alpha': 0.2})
```

```python
X_train = X_train[variables_name]
X_test = X_test[variables_name]
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.ensemble import RandomForestClassifier

#LogisticRegresion
lr = LogisticRegression(penalty='l2')
lr.fit(X_train_std, y_train)
y_pred_lr = lr.predict(X_test_std)

#SVM
svm = SVC(C=1.0)
svm.fit(X_train_std, y_train)
y_pred_svm = svm.predict(X_test_std)

#KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_std, y_train)
y_pred_knn = knn.predict(X_test_std)

#RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50)
rfc.fit(X_train, y_train)
y_pred_rfc = rfc.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler


def cross_valition_score(estimators, X_train, y_train):
    
    stdsc = StandardScaler()
    X_train_std = stdsc.fit_transform(X_train)
    X_test_std = stdsc.fit_transform(X_test)
    
    for estimator in estimators:
        
        if estimator is [lr, svm, knn]:
            scores = cross_val_score(estimator=estimator, X=X_train_std, y=y_train,
                                     cv=10, scoring='accuracy' , n_jobs=-1)
            print("Cross validation score (k=10) for {}: %.3f".format(estimator.__class__.__name__)  % scores.mean())
        else:
            scores = cross_val_score(estimator=estimator, X=X_train, y=y_train, cv=10, 
                                     scoring='accuracy' , n_jobs=-1)
            print("Cross validation score (k=10) for {}: %.3f".format(estimator.__class__.__name__)  % scores.mean())

estimators = [lr, svm, knn, rfc]

cross_valition_score(estimators, X_train_std, y_train)
```

```python
from sklearn.model_selection import learning_curve

def learning_curves(estimator,X_train, y_train):

    train_sizes, train_scores, test_scores = learning_curve(estimator=estimator, X=X_train, y=y_train, 
                                                            train_sizes=np.linspace(0.1, 1.0, 10),
                                                            cv=10, n_jobs=1)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.plot(train_sizes, train_mean, color='red', marker='o', markersize=5, label='Traning score')

    plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='red')

    plt.plot(train_sizes, test_mean, color='green', linestyle='--', marker='s', markersize=5, 
             label='Cross-validation score')

    plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')

    plt.grid()
    plt.xlabel('Traning examples')
    plt.ylabel('Score')
    plt.legend(loc='lower right')
    plt.ylim([0.8, 1.03])
    plt.title(estimator.__class__.__name__)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10), sharex = True, sharey = True)

ax = plt.gca()
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
fig.tight_layout(pad=6.0)


for i, estimator in zip(range(0, len(estimators)), estimators):
    
    stdsc = StandardScaler()
    X_train_std = stdsc.fit_transform(X_train)
    X_test_std = stdsc.fit_transform(X_test)
    
    fig.add_subplot(2, 2, i+1)
    
    if estimator is [lr, svm, knn]:
        learning_curves(estimator, X_train_std, y_train)
    else:
        learning_curves(estimator, X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV


#Logistic regression

param_grid_lr = [
    {'penalty':['l1','l2'], 'C':[0.001, 0.01, 0.1, 1.0, 10, 100]}
]

gs_lr = GridSearchCV(estimator=lr,
                  param_grid=param_grid_lr, 
                  scoring='accuracy', 
                  cv=10,
                  n_jobs=-1)

gs_lr = gs_lr.fit(X_train, y_train)
print('Best score for Logistic Regression: ' ,gs_lr.best_score_)
print('Best params for Logistic Regression: ', gs_lr.best_params_)


#SVM


param_grid_svm = [
    {'kernel':['linear', 'poly', 'rbf', 'sigmoid'], 'C':[0.001, 0.01, 0.1, 1.0, 10, 100]}
]

gs_svm = GridSearchCV(estimator=svm,
                  param_grid=param_grid_svm, 
                  scoring='accuracy', 
                  cv=10,
                  n_jobs=-1)

gs_svm = gs_svm.fit(X_train, y_train)
print('Best score for SVM: ',gs_svm.best_score_)
print('Best params for SVM', gs_svm.best_params_)



#KNN

param_grid_knn = [{'n_neighbors':[3 , 5, 7], 'metric':['euclidean', 'manhattan', 'chebyshev', 'minkowski']}]


gs_knn = GridSearchCV(estimator=knn,
                  param_grid=param_grid_knn, 
                  scoring='accuracy', 
                  cv=10,
                  n_jobs=-1)

gs_knn = gs_knn.fit(X_train, y_train)
print('Best score for KNN:',gs_knn.best_score_)
print('Best params for KNN:', gs_knn.best_params_)

#Random forest 
param_grid_rfc = [{'n_estimators':[10, 20, 30, 40, 50], 'max_features':[2, 4, 6, 8], 'max_depth' : [4,5,6]}]

gs_rfc = GridSearchCV(estimator=rfc,
                  param_grid=param_grid_rfc, 
                  scoring='accuracy', 
                  cv=10,
                  n_jobs=-1)

gs_rfc = gs_rfc.fit(X_train, y_train)
print('Best score for Random Forest: ', gs_rfc.best_score_)
print('Best params for Random Forest: ', gs_rfc.best_params_)

```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

best = [gs_lr, gs_svm, gs_knn, gs_rfc]
predictions = []



for classifier in best:
    clf = classifier.best_estimator_
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    predictions.append(y_pred)


for estimator, pred in zip(estimators, predictions):
    print("Accuracy score of {}: %.3f".format(estimator.__class__.__name__) % accuracy_score(pred, y_test))
    print("Confusion matrix of {}: \n".format(estimator.__class__.__name__), confusion_matrix(pred, y_test))
    print(" ") 
```

```python
import tensorflow as tf
from tensorflow import keras 
```

```python
from sklearn.preprocessing import MinMaxScaler

MinMaxS = MinMaxScaler()
X_train_mms = MinMaxS.fit_transform(X_train)
X_test_mms = MinMaxS.fit_transform(X_test)

from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X_train_mms, y_train, train_size = 0.9, random_state=42)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(10, activation='relu', input_shape=X_train_mms.shape[1:]))
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",
              optimizer='adam',
              metrics=["accuracy"])
```

```python
history = model.fit(np.asarray(X_train), np.asarray(y_train), epochs=10,
                    validation_data=(np.asarray(X_valid), np.asarray(y_valid)))
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)
```

```python
model.evaluate(np.array(X_test), np.array(y_test))
```
